## Estatísticas de Dados

### Variáveis qualitativas

```python
df_dict.query('tipo == "qualitativa"').variavel.to_list()
```

```python
df[['sex', 'smoker', 'day', 'time']].mode()
```

```python
quali_vars = ['sex', 'smoker', 'day', 'time']

for variavel in quali_vars:
    valores_unicos = df[variavel].unique()
    for valor in valores_unicos:
        print(f"{variavel} - {valor}")
        subamostra = df[df[variavel] == valor]
        display(subamostra.describe())
        print("="*100)
```

## Variáveis quantitativas


```python
quant_vars = ['total_bill', 'tip', 'size']

media = df[quant_vars].mean()
mediana = df[quant_vars].min()
desvio_padrao = df[quant_vars].std()

print("Média")
display(media)
print()
print("Mediana")
display(mediana)
print()
print("Desvio Padrão")
display(desvio_padrao)
```
